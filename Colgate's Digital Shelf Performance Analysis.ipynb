* In order to change Google sheet URL to CSV file to be understandble for pandas library : 

Replace /edit#gid={sheet_id} with /export?format=csv&gid={sheet_id}.

```python
# Function to convert google sgheet  URL to CSV format


def convert_google_sheet_url(sheet_id , spreadsheet_id):
    
    url_dict = {}
    
    for key in sheet_id : 
        url_dict[key] = f'https://docs.google.com/spreadsheets/d/{spreadsheet_id}/export?format=csv&gid={sheet_id[key]}'

    return url_dict
```

```python
# Making Dictionary of CSV for reading by pandas 

Url = convert_google_sheet_url(sheet_id= sheet_id , spreadsheet_id= spread_sheet_id)
```

```python
# Function for make dictionary of dataframe from each sheet on google sheet in order to manipulate them easier 

def convert_csv_dataframe (Url_Dict):
    
    sheet_df_dict = {}
    
    for key in Url_Dict:
        sheet_df_dict[key] = pd.read_csv(Url_Dict[key])
    
    return sheet_df_dict
```

```python


# Set seaborn style
sns.set(style="whitegrid")

# Plot the changes in various features
plt.figure(figsize=(16, 10))

# Define the list of features
features = ['Total Score', 'Total Score (Change)', 
            'Assortment & Availability Score', 'Assortment & Availability Score (Change)', 
            'Ratings & Reviews Score', 'Ratings & Reviews Score (Change)', 
            'Placement Score', 'Placement Score (Change)']

# Plot each feature separately
for i, feature in enumerate(features, start=1):
    plt.subplot(2, 4, i)
    sns.barplot(x='Category', y=feature+'_Q1', data=merged_df_ret2, color='skyblue', label='Q1')
    sns.barplot(x='Category', y=feature+'_Q4', data=merged_df_ret2, color='lightgreen', label='Q4')
    plt.xlabel('Category')
    plt.ylabel(feature)
    plt.title(f'{feature} by Category (Retailer 2)',fontsize = 13 , rotation = 5)
    plt.xticks(rotation=45)
    plt.legend()

plt.tight_layout()
plt.savefig(r'C:\Users\Arvin\Desktop\Assignment\Q1_Q4\Quality_increased\All Feature bar  (Retailer 2).png', dpi = 300)
plt.show()
```
